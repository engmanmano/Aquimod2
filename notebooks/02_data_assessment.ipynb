# Objective A: Groundwater Availability and Variability Assessment

This notebook supports **Objective A** of Enggy's thesis:

> **Assess current groundwater availability and variability in selected key areas of the Cagayan River Basin using available observation records and historical datasets.**

---

### Purpose
This notebook is designed to:
- Load and clean historical groundwater level data
- Visualize temporal and seasonal trends in groundwater availability
- Summarize patterns and prepare data insights for model calibration

---

### Input Requirements
- Observation records in CSV or text format
- At minimum, the data should include:
  - Date (or separate Day, Month, Year)
  - Groundwater Level (GWL)
  - Optional: Rainfall, PET, Soil Moisture

---

### Outputs
- Time series plots of GWL
- Seasonal boxplots showing variability
- Monthly summary statistics
- Cleaned dataset ready for simulation or further analysis

---

> *Note: This notebook is exploratory and will be updated as observation data becomes available.*
